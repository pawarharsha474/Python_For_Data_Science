```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'A':['a0','a1','a2','a3'],
                  'B':['b0','b1','b2','b3'] , 
                  'C':['c0','c1','c2','c3'] ,
                  'D':['d0','d1','d2','d3'],
                   'index':[0,1,2,3]})
```

```python
df2=pd.DataFrame({'A':['a4','a5','a6','a7'],
                  'B':['b4','b5','b6','b7'] , 
                  'C':['c4','c5','c6','c7'] ,
                  'D':['d4','d5','d5','d7'],
                   'index':[4,5,6,7]})
```

```python
df3=pd.DataFrame({'A':['a8','a9','a10','a11'],
                  'B':['b8','b9','b10','b11'] , 
                  'C':['c8','c9','c10','c11'] ,
                  'D':['d8','d9','d10','d11'],
                   'index':[8,9,10,11]})
```

```python
result=pd.concat([df1, df2, df3])
result
```

```python
#combining dataframes with keys
result=pd.concat([df1, df2, df3], keys=['x','y','z'])
result
```

```python
#get 2nd dataframe
result.loc['y']
```

# concating using axis

```python
df4=pd.DataFrame({'B':['b2','b3','b6','b7'],
                  'D':['d2','d3','d6','d7'] , 
                  'F':['f2','f3','f6','f7'] ,
                   'index':[2,3,6,7]})
```

```python
df4
```

```python
result=pd.concat([df1,df4],axis=1,sort=False)
result
```

```python
result=pd.concat([df1,df4],axis=1,join='inner')
result
```

```python
result=pd.concat([df1,df4],axis=1,join='outer')
result
```
